## wartości kategorialne

```python
import pandas as pd
import numpy as np
```

```python
# dane z brakami
X = pd.DataFrame([['M', 2.15, 5.45],
                 ['F', 2.27, 3.53],
                 ['F', 1.01, -0.56],
                 ['M', -0.53, 0.39],
                 ['M', 3.65, 5.21],
                 [np.nan, 1.68, 5.23],
                 [np.nan, -0.35, 2.56]])
```

```python
# 8. Sprawdzić statystyki brakujących wartości

X.isnull().sum()
```

```python
# 9. Imputacja za pomocą klasy SimpleImputer
#...
# X_imputed1 = 
simp_imp = SimpleImputer(strategy='most_frequent')
X_imputed1 = pd.DataFrame(simp_imp.fit_transform(X), columns=X.columns)
X_imputed1
```

```python
# 10. Imputacja za pomocą klasy KNNImputer
#...
# X_imputed2 = 
knn_imp2 = KNNImputer(n_neighbors=7)
X_imputed2 = pd.DataFrame(knn_imp2.fit_transform(X), columns=X.columns)
```
